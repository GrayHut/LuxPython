>Data cleaning project with pandas - Hotel booking data from kaggle(csv)

```python
#importing pandas

import pandas as pd
```

```python
#read data from the dataset

hotel_data = pd.read_csv('./csvs/hotel_bookings_dirty.csv')
```

```python
#first look at the dataset
hotel_data.head()   #displays the first five rows of data in the table

hotel_data  #displays the csv table
```

```python
#dropping and naming columns

hotel_data.columns  #displays the columns of the table
```

```python
hotel_data.drop(columns='required_car_parking_spaces', inplace=True)    #deletes the column in question
```

```python
#renaming of select columns

hotel_data.rename({'adr': 'address', 'adults': 'num_adults', 'children': 'num_children', 'babies' : 'num_babies'}, axis = 1, inplace = True)
```

```python
hotel_data.head(10) #display of the first ten rows
```

```python
#Nans >> NULLs

#calculate percentage of NULL/NaN values within each column
#both the below lines of code achieve the same output

hotel_data.isnull().sum() * 100 / len(hotel_data)

(hotel_data.isnull().sum() / len(hotel_data))* 100
```

```python
#replacing NaN/NULL values in a column
hotel_data['agent'].fillna(-1, inplace=True)
hotel_data['company'].fillna(0, inplace=True)
```

```python
#checking for null values
hotel_data[hotel_data['num_children'].isna()]

hotel_data[hotel_data['country'].isna()][['num_adults', 'booking_changes', 'country']]
```

```python
#dropping column NULL values
hotel_data.dropna(subset=['num_children'], inplace=True)

#dropping entire column
hotel_data.drop(columns=['company'], inplace=True)
```

```python
#checking column datatypes
hotel_data.dtypes
```

```python
#changing column types
hotel_data = hotel_data.astype({'num_children': 'int', 'is_canceled': 'boolean', 'is_repeated_guest': 'boolean'})
```

```python
#bin columns

#display column unique values
hotel_data['lead_time'].unique

#gives a summary break-down of the column values
hotel_data['lead_time'].describe()  

bin = [0, 250, 500, 750]
label = ['0-250', '250-500', '500-750']

hotel_data['lead_time_binned'] = pd.cut(hotel_data['lead_time'], bins=bin, labels=label)
hotel_data[['lead_time', 'lead_time_binned']]
```

```python
#separation of columns

hotel_data['arrival_date_month'] = hotel_data['arrival_date'].str.split('-', expand=True)[0]
hotel_data['arrival_date_year'] = hotel_data['arrival_year'].str.split('-', expand=True)[1]
```

```python
#moving columns
column_to_move = hotel_data.pop('arrival_date_month')
hotel_data.insert(3, 'arrival_date_month', column_to_move)
hotel_data.head()
```

```python
#string cleaning

hotel_data['hotel'].unique()

# removes any single occurrence of *, a newline, or ^
hotel_data['hotel'] = hotel_data['hotel'].replace(r"[\*\n\^]", '', regex=True)
```

```python
#removing duplicates

hotel_data.duplicated(keep=False)

hotel_data.loc[hotel_data.duplicated(keep=False)]

```

```python
hotel_data.drop_duplicates(keep="first", inplace=True)
```
